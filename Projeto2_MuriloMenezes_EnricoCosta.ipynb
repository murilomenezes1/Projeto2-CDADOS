### Escolha de um produto e coleta das mensagens


```python
#Produto escolhido:
produto = 'FIFA 20'

#Quantidade m√≠nima de mensagens capturadas:
n = 750
#Quantidade m√≠nima de mensagens para a base de treinamento:
t = 500

#Filtro de l√≠ngua, escolha uma na tabela ISO 639-1.
lang = 'pt'
```

Capturando os dados do twitter:

```python
#Cria um objeto para a captura
api = tweepy.API(auth)

#Inicia a captura, para mais detalhes: ver a documenta√ß√£o do tweepy
i = 1
msgs = []
for msg in tweepy.Cursor(api.search, q=produto, lang=lang, tweet_mode="extended").items():
    
    if msg.full_text.lower()[0] != 'r' and msg.full_text.lower()[1] != 't':
    
        msgs.append(msg.full_text.lower())
        i += 1
        if i > n:
            break

#Embaralhando as mensagens para reduzir um poss√≠vel vi√©s
shuffle(msgs)
```

Salvando os dados em uma planilha Excel:

```python
#Verifica se o arquivo n√£o existe para n√£o substituir um conjunto pronto
if not os.path.isfile('./{0}-2.xlsx'.format(produto)):
    
    #Abre o arquivo para escrita
    writer = pd.ExcelWriter('{0}-2.xlsx'.format(produto))

    #divide o conjunto de mensagens em duas planilhas
    dft = pd.DataFrame({'Treinamento' : pd.Series(msgs[:t])})
    dft.to_excel(excel_writer = writer, sheet_name = 'Treinamento', index = False)

    dfc = pd.DataFrame({'Teste' : pd.Series(msgs[t:])})
    dfc.to_excel(excel_writer = writer, sheet_name = 'Teste', index = False)

    #fecha o arquivo
    writer.save()
```

### Montando o Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
import re
def cleanup(text):
   
    import string
    punctuation = '[!-.:?;&]' 
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, ' ', text)
    return text_subbed
```

```python
twitter_treinamento = pd.read_excel("FIFA 20-2.xlsx",sheet_name="Treinamento")
twitter_teste = pd.read_excel("FIFA 20-2.xlsx",sheet_name="Teste")

twitter_teste["Teste s/ pont."] = twitter_teste["Teste"].apply(cleanup)
twitter_teste=twitter_teste.drop(columns="Teste")

twitter_teste["Teste s/ pont."]=twitter_teste["Teste s/ pont."].str.lower()

twitter_treinamento["Treinamento s/ pont."] = twitter_treinamento["Treinamento"].apply(cleanup)
twitter_treinamento=twitter_treinamento.drop(columns="Treinamento")

twitter_treinamento["Treinamento s/ pont."]=twitter_treinamento["Treinamento s/ pont."].str.lower()


```

```python
treinamento_muito_irrelevante = twitter_treinamento.loc[twitter_treinamento.Classificacao==0]
treinamento_irrelevante = twitter_treinamento.loc[twitter_treinamento.Classificacao==1]
treinamento_neutro = twitter_treinamento.loc[twitter_treinamento.Classificacao==2]
treinamento_relevante = twitter_treinamento.loc[twitter_treinamento.Classificacao==3]
treinamento_muito_relevante = twitter_treinamento.loc[twitter_treinamento.Classificacao==4]
```

```python
Muito_Irrelevante = []
for i in treinamento_muito_irrelevante["Treinamento s/ pont."]:
    Muito_Irrelevante.append(i)
    
MI = ','.join(Muito_Irrelevante)

serie_treinamento_MI = pd.Series(MI.split())
tabela_treinamento_MI= serie_treinamento_MI.value_counts()
tabela_treinamento_MI_relativa = serie_treinamento_MI.value_counts(True)

Irrelevante = []
for i in treinamento_irrelevante["Treinamento s/ pont."]:
    Irrelevante.append(i)

I = ','.join(Irrelevante)

serie_treinamento_I = pd.Series(I.split())
tabela_treinamento_I = serie_treinamento_I.value_counts()
tabela_treinamento_I_relativa = serie_treinamento_I.value_counts(True)


Neutro = []
for i in treinamento_neutro["Treinamento s/ pont."]:
    Neutro.append(i)

N = ','.join(Neutro)

serie_treinamento_N = pd.Series(N.split())
tabela_treinamento_N = serie_treinamento_N.value_counts()
tabela_treinamento_N_relativa = serie_treinamento_N.value_counts(True)

Relevante = []
for i in treinamento_relevante["Treinamento s/ pont."]:
    Relevante.append(i)
    
R = ','.join(Relevante)

serie_treinamento_R = pd.Series(R.split())
tabela_treinamento_R = serie_treinamento_R.value_counts()
tabela_treinamento_R_relativa = serie_treinamento_R.value_counts(True)

Muito_Relevante = []
for i in treinamento_muito_relevante["Treinamento s/ pont."]:
    Muito_Relevante.append(i)

MR = ','.join(Muito_Relevante)

serie_treinamento_MR = pd.Series(MR.split())
tabela_treinamento_MR = serie_treinamento_MR.value_counts()
tabela_treinamento_MR_relativa = serie_treinamento_MR.value_counts(True)

dataset = MI+I+N+R+MR

serie_treinamento_ds = pd.Series(dataset.split())
tabela_treinamento_ds = serie_treinamento_ds.value_counts()
tabela_treinamento_ds_relativa = serie_treinamento_ds.value_counts(True)

twitter_treinamento["Classificacao"].value_counts()
```

```python
#Interse√ß√£o entre as classifica√ß√µes

set_MI = set(tabela_treinamento_MI.index)
set_I = set(tabela_treinamento_I.index)
set_N = set(tabela_treinamento_N.index)
set_R = set(tabela_treinamento_R.index)
set_MR = set(tabela_treinamento_MR.index)

inter1 = set_MI.intersection(set_I)
inter2 = set_N.intersection(inter1)
inter3 = set_R.intersection(inter2)
inter_total = set_MR.intersection(inter3)




```

```python
# Calculo das probabilidades

teste = []

Tteste = twitter_teste["Teste s/ pont."]

Tteste
for i in Tteste:
    teste.append(i.split())

    
prob_R=[]
prob_I=[]
prob_N=[]
prob_MR=[]
prob_MI=[]


for tweet in teste:
    
    pr = (74/500)*(1/(len(tabela_treinamento_R)+len(tabela_treinamento_ds)))
    pi = (131/500)*(1/(len(tabela_treinamento_I)+len(tabela_treinamento_ds)))
    pn = (230/500)*(1/(len(tabela_treinamento_N)+len(tabela_treinamento_ds)))
    pmr = (29/500)*(1/(len(tabela_treinamento_MR)+len(tabela_treinamento_ds)))
    pmi = (36/500)*(1/(len(tabela_treinamento_MI)+len(tabela_treinamento_ds)))
    
    for palavra in tweet:
        
        if palavra in tabela_treinamento_MI:
            
            pmi = pmi*tabela_treinamento_MI_relativa[palavra]
        
        else: 
            cont_MI = 1 
            pmi = pmi*(cont_MI/(len(tabela_treinamento_MI)+len(tabela_treinamento_ds)))
            
        
        if palavra in tabela_treinamento_N:
            
            pn = pn*tabela_treinamento_N_relativa[palavra]
        
        else:
            cont_N = 1
            pn = pn*(cont_N/(len(tabela_treinamento_N)+len(tabela_treinamento_ds)))
            
                       
        if palavra in tabela_treinamento_R:
            
            pr = pr*tabela_treinamento_R_relativa[palavra]
        
        else: 
            cont_R = 1 
            pr = pr*(cont_R/(len(tabela_treinamento_R) + len(tabela_treinamento_ds)))
                     
        if palavra in tabela_treinamento_MR:
        
             pmr = pmr*tabela_treinamento_MR_relativa[palavra]
                     
        else:

            cont_MR = 1
            pmr = pmr*(cont_MR/(len(tabela_treinamento_MR)+len(tabela_treinamento_ds)))
            
            
        if palavra in tabela_treinamento_I:
            pi = pi*tabela_treinamento_I_relativa[palavra]
        else:
            cont_I = 1 
            pi = pi*(cont_I/(len(tabela_treinamento_I)+len(tabela_treinamento_ds)))
                       
        prob_R.append(pr)
        prob_MI.append(pmi)
        prob_MR.append(pmr)
        prob_N.append(pn)
        prob_I.append(pi)
        
twitter_teste["Verifica√ß√£o"] =-1

for i in range(0,250):
    
    if prob_MR[i] > prob_R[i] and prob_MR[i] > prob_N[i] and prob_MR[i] > prob_I[i] and prob_MR[i] > prob_MI[i]:
        twitter_teste.iloc[i,2] = 4
    if prob_R[i] > prob_MR[i] and prob_R[i] > prob_N[i] and prob_R[i] > prob_I[i] and prob_R[i] > prob_MI[i]:
        twitter_teste.iloc[i,2] = 3
    if prob_N[i] > prob_MR[i] and prob_N[i] > prob_R[i] and prob_N[i] > prob_I[i] and prob_N[i] > prob_MI[i]:
        twitter_teste.iloc[i,2] = 2
    if prob_I[i] > prob_MR[i] and prob_I[i] > prob_R[i] and prob_I[i] > prob_N[i] and prob_I[i] > prob_MI[i]:
        twitter_teste.iloc[i,2] = 1
    if prob_MI[i] > prob_MR[i] and prob_MI[i] > prob_R[i] and prob_MI[i] > prob_N[i] and prob_MI[i] > prob_I[i]:
        twitter_teste.iloc[i,2] = 0
        


    

        


```

### Concluindo

A partir dos resultados obtidos, onde o percentual de acerto do classificador √© de 37,9%, pode se concluir que existe a necessidade de um n√∫mero maior de tweets. Tal necessidade decorre do fato de que, com mais frases classificadas, o treinamento ir√° ficar cada vez mais aprimorado e, como consequ√™ncia disso, a verifica√ß√£o estar√° cada vez mais pr√≥xima da classifica√ß√£o previamente feita.

### por que n√£o posso usar o pr√≥prio classificador para gerar mais amostras de treinamento

O classificador feito n√£o deve ser usado para gerar mais amostras de treinamento pois este precisa ser uma base correta e definida. Sendo assim, n√£o √© pr√≥prio utilizar novas amostras para colocar no treinamento pois, com a taxa de erro, o treinamento ir√° ficar incorreto e assim tweets classificados errados ir√£o se tornar padr√£o.

### Melhorias para este classificador

Para que o projeto fosse melhorado, apresentando uma classificador melhor e com uma probabilidade maior de acerto, seria necess√°rio obter entre 1000 e 1500 tweets, para que assim seja feita uma an√°lise levando em considera√ß√£o cinco categorias ( Muito relevante, Relevante, Neutro, Irrelevante, Muito Irrelevante) que n√£o apresente grande dispers√£o na m√©dia de tweets por categoria. Desta maneira, o classificador teria um treinamento apropriado e conseguiria prever com maior exatid√£o a probabilidade de um novo tweet pertencer a determinada categoria.
